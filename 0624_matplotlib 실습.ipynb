```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')
plt.show()
```

```python
x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)
tsin_y = np.sin(2*x)
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')
ax3.plot(x, tan_y, 'g--')
ax4.plot(x, tsin_y, 'y--')
plt.show()
```
