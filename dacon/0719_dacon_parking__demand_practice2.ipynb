### 컬럼명 변경

### 임대보증금 '-' 인 값을 null값으로 바꿔주고 
### (임대보증금,임대료) type변경 -> float

```python
train.loc[train.임대보증금=='-', '임대보증금'] = np.nan
test.loc[test.임대보증금=='-', '임대보증금'] = np.nan
train['임대보증금'] = train['임대보증금'].astype(float)
test['임대보증금'] = test['임대보증금'].astype(float)

train.loc[train.임대료=='-', '임대료'] = np.nan
test.loc[test.임대료=='-', '임대료'] = np.nan
train['임대료'] = train['임대료'].astype(float)
test['임대료'] = test['임대료'].astype(float)
```

### (임대보증금, 임대료) 결측값 -> 0으로 대체

```python
train[['임대보증금', '임대료']] = train[['임대보증금', '임대료']].fillna(0)
test[['임대보증금', '임대료']] = test[['임대보증금', '임대료']].fillna(0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
#'총세대수', '전용면적', '전용면적별세대수','임대건물구분','지역',
#'공급유형', '공가수','지하철수','자격유형', '버정수', '임대보증금','임대료','단지내주차면수'
sel = [ '총세대수','전용면적별세대수', '전용면적','임대건물구분','지역','공급유형', '공가수',
       '지하철수','자격유형', '버정수', '임대보증금','임대료','단지내주차면수']

X_train = train[sel]
y_train = train['등록차량수']
X_test = test[sel]
```

### 선형회귀

```python
model = LinearRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
```

### 라쏘적용

```python
model = Lasso()
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
```

```python
alpha_p = [0, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 5, 10]

for p in alpha_p:
    lasso = Lasso(alpha=p).fit(X_train, y_train)
    tr_score = lasso.score(X_train, y_train)
    print("alpha : {} 학습 : {}".format(p,tr_score))
    print("유효한 feature 개수 : ", np.sum(lasso.coef_ !=0))
```

### 릿지적용

```python
model = Ridge(alpha=0.0001)
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
```

```python
alpha_p = [0, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 5, 10]

for p in alpha_p:
    ridge = Ridge(alpha=p).fit(X_train, y_train)
    tr_score = ridge.score(X_train, y_train)
    print("alpha : {} 학습 : {}".format(p,tr_score))
```
