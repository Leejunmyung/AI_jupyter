```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

- 버전확인

```python
x = np.linspace(0,14,100)
y = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 4*np.sin(x+1.0)

plt.figure(figsize=(10,6)) # 그림의 크기
plt.plot(x,y)
```

```python
x = np.linspace(0,14,100) #0부터 14까지 동일한 간격의 값 100개 만들기
x
```

```python
y1 = np.sin(x)
y2 = np.sin(x)*2
y3 = np.sin(x)*4
plt.plot(x,y1, x,y2, x,y3)
```

```python
## 그래프에 범례 표시해보기
plt.figure(figsize=(8,6))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="sin(x)*2")
plt.plot(x, y3, label="sin(x)*4")
plt.legend()
plt.title("sin plot")
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.head()
```

```python
tips.tail()
```

```python
### 요일별 식사 금액은 얼마나 될까?
plt.figure(figsize=(8,6))
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Amount of meal")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="day", y="tip", hue='sex', data=tips)
plt.title("Tips by day")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="tip", hue='sex', data=tips)
plt.title("Tips by day")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, size=7)
plt.show()
```

```python
import seaborn as sns
sns.relplot(x="tip", y="total_bill", hue="smoker", size='size', #alpha 선명도
            sizes=(40,400), alpha=.8,palette="muted",height=7, data=tips) #palette 색바꿔줌
```

```python
fg = sns.load_dataset('flights')
print(fg.shape    ) #행과 열
print(fg.columns  ) #컬럼명 확인
print(fg.info()   ) #행과열, 컬럼명, 자료형
print(fg.dtypes   ) #컬럼의 자료형
print(fg.describe())#수치형의 요약값
```

```python
fgp = fg.pivot("month", "year", "passengers")
fgp
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(fgp, annot=True, fmt="d")
plt.show()
```
