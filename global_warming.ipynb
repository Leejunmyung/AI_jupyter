```python
df = pd.read_csv(
                "gt.csv",
                usecols=['Country','dt','AverageTemperature'],
                parse_dates=['dt'])
countries = ['United States', 'India', 'China', 'Russia','Brazil']
df = df[df['Country'].isin(countries)]
df = df[df['dt'] > '2000-01-01']
print(df)
import matplotlib as mpl
from matplotlib import pyplot as plt
data = df.pivot('dt', 'Country', 'AverageTemperature')
data=data.fillna(0)
print(data)
data.plot()
```
