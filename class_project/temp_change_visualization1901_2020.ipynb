```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.offline as py
py.init_notebook_mode(connected=True)

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

global_temp_country = pd.read_csv('tas_1901_2020.csv')
```

```python
#컬럼명 변경
global_temp_country.columns=['temp','year','month','country']
global_temp_country.head()
# print(type(global_temp_country))
# print(type(global_temp_country['month']))

#month 정보 뽑기 : 앞뒤 공백 제거후 slicing
global_temp_country['month'] = global_temp_country['month'].str.strip().str.split(' ',expand = True)[0] 
global_temp_country.head()

#나라별 연도별 평균 기온 구하기
temp_avg = global_temp_country[['country','year','temp']].groupby(['year','country']).mean().reset_index()

```

```python
px.choropleth(data_frame=temp_avg,locations="country",locationmode='country names',
              animation_frame="year",color='temp',
              color_continuous_scale = 'Turbo',
              title="1901년부터 2020년까지 국가별 평균 기온변화")
```
