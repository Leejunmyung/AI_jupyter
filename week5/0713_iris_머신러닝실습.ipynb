## 3-2 tips을 이용해서 knn 모델을 구해보자.
 * X : total_bill, tip
 * y : size
 * 02 데이터 나누기
 * 03 knn 모델 만들기
 * 04 평가해 보기

```python
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
```

```python
from sklearn.model_selection import train_test_split

sel = ['total_bill', 'tip']
X = tips[sel]
y = tips['size']

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=33)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
pred
```

```python
knn.score(X_test, y_test)
```

```python
np.mean(pred == y_test)
```

```python
for i in range(1,50,1):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    print(i,knn.score(X_test, y_test))
```

```python
### 좋은 k를 찾기
good_k = 999 # 좋은 k찾기
acc = 0      # 정확도
for i in range(1,50,1):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    if acc < np.mean(pred==y_test):
        good_k = i                   # k값 업데이트
        acc = np.mean(pred==y_test)  # 정확도 업데이트
        print(good_k, np.mean(pred==y_test))
```

### 3-4 titanic의 survived를 예측하는 모델(knn) 모델을 만들어보자.
* 데이터 나누기
* knn 모델
* 평가

```python
titanic
```

```python
from sklearn.model_selection import train_test_split

sel = ['pclass', 'sibsp','parch']
X = titanic[sel]
y = titanic['survived']

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
acc = 0
for i in range(1,50,1):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    if acc < np.mean(pred==y_test):
        acc = np.mean(pred==y_test)
        print(i,acc)
```
