```python
train = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')
sub = pd.read_csv('/kaggle/input/titanic/gender_submission.csv')

train.shape, test.shape, sub.shape
```

```python
train['Embarked'] = train['Embarked'].fillna('S')
train['Embarked'].isnull().sum()
```

```python
test.Fare.value_counts()
```

```python
test['Fare'] = test['Fare'].fillna(test.Fare.mean())
```

```python
test['Fare'].isnull().sum()
```

```python
train['Sex'] = train['Sex'].map({'male':0, 'female':1}).astype(int)
test['Sex'] = test['Sex'].map({'male':0, 'female':1}).astype(int)
```

```python
train['Embarked'] = train['Embarked'].map({'S':0, 'C':1, "Q":2}).astype(int) ##자료형 변경
test['Embarked'] = test['Embarked'].map({'S':0, 'C':1, "Q":2}).astype(int) ##자료형 변경
```

```python
train.info()
```

```python
test.info()
```

```python
train.loc[train['Cabin'].isna(), 'Cabin_bool'] = 1
train.loc[~train['Cabin'].isna(), 'Cabin_bool'] = 0
```

```python
train.head()
```

```python
test.head()
```

```python
test.loc[train['Cabin'].isna(), 'Cabin_bool'] = 1
test.loc[~train['Cabin'].isna(), 'Cabin_bool'] = 0
```

```python
test.head()
```

```python
sub
```

```python
all_df = [train, test]
for one in all_df:
    one['Title'] = one.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
test.head()
```

```python
train['Title'].value_counts()
```

```python
for dataset in all_df:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt','Col',
                                                'Don','Dr','Major','Rev',
                                                'Sir','Jonkheer','Dona'],'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle','Miss')
    dataset['Title'] = dataset['Title'].replace('Ms','Miss')
    dataset['Title'] = dataset['Title'].replace('Mme','Mrs')
    
```

```python
test['Title'].value_counts()
```

```python
train['Title'].value_counts()
```

```python
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
for dataset in all_df:
    dataset['Title'] = dataset['Title'].map(title_mapping)
```

```python
train['Family_size'] = train['SibSp'] + train['Parch'] + 1
test['Family_size'] = test['SibSp'] + test['Parch'] + 1
```

```python
train
```

```python
all_titanic = pd.concat([train, test], join='inner')
all_titanic.shape, all_titanic.columns
```

```python
all_titanic['Age'] = all_titanic['Age'].fillna( all_titanic['Age'].mean()  )
all_titanic.info()
```

```python
all_titanic.head()
```

```python
all_titanic['qcut_Age'] = pd.qcut(all_titanic['Age'], 5, labels=False)
all_titanic
```

```python
train_df = all_titanic.iloc[0:891, :]
test_df = all_titanic.iloc[891:, :]
train_df.shape, test_df.shape
```

```python
train_df = pd.concat([train_df, train['Survived']],axis=1 )
```

```python
train_df.head()
```

```python
sel = ['Pclass', 'SibSp', 'Parch', 'Sex','Embarked','Cabin_bool','Title','Family_size']
X_train = train_df[sel]
X_test = test_df[sel]
y_train = train_df['Survived']
```

```python
from sklearn.tree import DecisionTreeClassifier
decisiontree = DecisionTreeClassifier()
decisiontree.fit(X_train, y_train)
# 예측
pred = decisiontree.predict(X_test)
pred[:15]
```

```python
sub['Survived'] = pred
sub.to_csv("decision_four_model.csv", index=False)
```
