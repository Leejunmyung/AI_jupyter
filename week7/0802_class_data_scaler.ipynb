```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
print("[cancer.keys()]  \n{}".format(cancer.keys()))
print("   : {}".format(cancer.data.shape))
```

```python
#   
X_train, X_test, y_train, y_test = train_test_split(cancer.data, 
                                                    cancer.target, 
                                                    stratify=cancer.target, 
                                                    random_state=77)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
import numpy as np
# np.set_printoptions(precision=2)
np.set_printoptions(formatter={'float_kind': lambda x: "{0:0.3f}".format(x)})
```

```python
X_train_scaled = scaler.transform(X_train)

print(" ,  : \n {}, {}".format(X_train.max(axis=0), 
                                      X_train.min(axis=0)))
print()
print(" ,  : \n {}, {}".format(X_train_scaled.max(axis=0), 
                                      X_train_scaled.min(axis=0)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.boxplot(X_train)
plt.subplot(1,2,2)
sns.boxplot(X_train_scaled)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#    
X_train_scaler = scaler.fit(X_train).transform(X_train)
X_train_scaler_d = scaler.fit_transform(X_train)
### fit().transform() fit_transform() .
print(" :", X_train.max(), X_train.min() )
print(" :", X_train_scaler.max(), X_train_scaler.min() )
print(" :", X_train_scaler_d.max(), X_train_scaler_d.min() )
```

```python
from sklearn.datasets import make_blobs

#   
X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)

#     
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)
```

```python
X.shape
```

```python
#      
fig, axes = plt.subplots(1, 3, figsize=(13, 4))
axes[0].scatter(X_train[:, 0], X_train[:, 1],
                c=mglearn.cm2(0), label=" ", s=60)
axes[0].scatter(X_test[:, 0], X_test[:, 1], marker='^',
                c=mglearn.cm2(1), label=" ", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title(" ")

# MinMaxScaler   
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

#     
axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
                c=mglearn.cm2(0), label=" ", s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], marker='^',
                c=mglearn.cm2(1), label=" ", s=60)
axes[1].set_title("  ")

#     
#    0,  1 
#        
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)

#     
axes[2].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
                c=mglearn.cm2(0), label="training set", s=60)
axes[2].scatter(X_test_scaled_badly[:, 0], X_test_scaled_badly[:, 1],
                marker='^', c=mglearn.cm2(1), label="test set", s=60)
axes[2].set_title("  ")

for ax in axes:
    ax.set_xlabel(" 0")
    ax.set_ylabel(" 1")
fig.tight_layout()
```
