```python
import seaborn as sns
import plotly
import cufflinks as cf
import pandas as pd
import numpy as np
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
titanic.age.value_counts()
```

```python
titanic.deck.value_counts()
```

```python
titanic.embark_town.value_counts()
```

```python
titanic['embark_town'] = titanic['embark_town'].fillna('Southampton')
titanic['embarked'] = titanic['embarked'].fillna('S')
titanic['deck'] = titanic['deck'].fillna('C')
titanic['age'] = titanic['age'].fillna(30)
titanic.isnull().sum()
```

```python
titanic
```

```python
titanic.iplot(kind='scatter', x='age', y='fare', 
              mode='markers', size=20, symbol=17, color='lightgreen')
```

```python
titanic.iplot(kind='bar', x='sex', y='survived')
```

```python
titanic.iplot(kind='bar', x='embark_town', y='survived')
```

```python
titanic.iplot(kind='bar', x='who',
             y=['sibsp','parch'], xTitle='who', yTitle='family')
```
