### 실습
 * sepal length, width의 행의 합을 구해보자.(sepal_sum 컬럼 생성하기)
 * petal length, width의 행의 합을 구해보자.(petal_sum 컬럼 생성하기)


```python
iris['sepal_sum'] = iris[['sepal_length','sepal_width']].sum(axis=1)
iris
```

```python
iris['petal_sum'] = iris[['petal_length','petal_width']].sum(axis=1)
iris
```

```python
iris.drop('all_sum', axis=1, inplace=True)
```

```python
iris.drop('sepal_sum',axis=1)
```

```python
iris.drop(0)
```

### 실습 3-2
 * length별 합계구하고, 컬럼 만들기
 * width별 합계 구하고, 컬럼 만들기
 * drop이용해서 2개 컬럼을 삭제해 보기

```python
iris['length_sum'] = iris[['sepal_length','petal_length']].sum(axis=1)
iris
```

```python
iris['width_sum'] = iris[['sepal_width','petal_width']].sum(axis=1)
iris
```

```python
iris.drop(['length_sum','width_sum'],axis=1)
```
