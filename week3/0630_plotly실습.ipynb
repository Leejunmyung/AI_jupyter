```python
iris.iplot(kind='bar', x='species', y='sepal_width')
```

### 실습 2-1
 * iris 데이터 셋의 sepal_width와 petal_length에 대해서 막대그래프 그려보기
 * x축, y축, 제목을 달아주세요.
 * 그래프를 보고 .. 한줄 코멘트 달아주기.

```python
iris.iplot(kind='bar', x='species', y='petal_length', 
           title='species별 petal_length 구하기',
           xTitle='species(종류별)', yTitle='petal_length')
```
